```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv(r'C:\Users\Kieron\PycharmProjects\Kaggle\Titanic\data\input\train.csv', header=0)
df_test = pd.read_csv(r'C:\Users\Kieron\PycharmProjects\Kaggle\Titanic\data\input\test.csv', header=0)
```

```python
Y_train = df_train['Survived']
```

```python
X_train = df_train.drop(['Survived'], axis=1)
```

```python
X_train.head()
```

```python
X_test = df_test.copy(deep=True)
```

```python
X_train.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
X_train_encoded = X_train.copy(deep=True)
```

```python
# 'Embarked' should be imputed here
X_train_encoded = pd.concat([X_train, (pd.get_dummies(X_train_encoded[['Sex', 'Embarked']]))], axis=1)
X_train_encoded = X_train_encoded.drop(['Sex', 'Embarked'], axis=1)
X_test = pd.concat([X_test, (pd.get_dummies(X_test[['Sex', 'Embarked']]))], axis=1)
X_test = X_test.drop(['Sex', 'Embarked'], axis=1)
```

```python
X_train_encoded.tail()
```

```python
X_test.tail()
```

```python
'''
Impute null data
'''
# Impute null
from sklearn.preprocessing import Imputer
imr_age = Imputer(missing_values='NaN', strategy='median', axis=0)
```

```python
# Fit datasets to these imputers & do imputation
# X_train
imr_age = imr_age.fit(X_train_encoded[['Age']])
X_train_encoded[['Age']] = imr_age.transform(X_train_encoded[['Age']].values)
# X_test
imr_age = imr_age.fit(X_test[['Age']])
X_test[['Age']] = imr_age.transform(X_test[['Age']].values)
```

```python
X_train_encoded.head()
```

```python
X_train_encoded.isnull().sum()
```

```python
# キャビンの頭文字（ランク？）
# X_train
X_train_encoded['cabin_char'] = X_train_encoded['Cabin'].str.extract('(\w)\d+', expand=False)
X_train_encoded['cabin_char'] = X_train_encoded['cabin_char'].fillna('N')
for c in list('ABCDEFGN'):
    X_train_encoded['cabin_' + c] = (X_train_encoded['cabin_char'] == c).astype(int)
# X_test
X_test['cabin_char'] = X_test['Cabin'].str.extract('(\w)\d+', expand=False)
X_test['cabin_char'] = X_test['cabin_char'].fillna('N')
for c in list('ABCDEFGN'):
    X_test['cabin_' + c] = (X_test['cabin_char'] == c).astype(int)
```

```python
# キャビンの番号 適当に区分化
# X_train
X_train_encoded['cabin_num'] = X_train_encoded['Cabin'].str.extract('\w(\d+)', expand=False).astype(float)
X_train_encoded['cabin_num'] = X_train_encoded['cabin_num'].fillna(75)  # 適当に中間
X_train_encoded['cabin_0_50'] = \
    np.where(X_train_encoded['cabin_num'] <= 50, 1, 0) # pd.get_dummies( )
X_train_encoded['cabin_50_100'] = \
    np.where((50 < X_train_encoded['cabin_num']) & (X_train_encoded['cabin_num'] <= 100), 1, 0) # pd.get_dummies( ) 
X_train_encoded['cabin_100_'] = \
    np.where(100 < X_train_encoded['cabin_num'], 1, 0)

# 客室数
X_train_encoded['n_cabin'] = X_train_encoded['Cabin'].str.count(' ') + 1

# X_test
X_test['cabin_num'] = X_test['Cabin'].str.extract('\w(\d+)', expand=False).astype(float)
X_test['cabin_num'] = X_test['cabin_num'].fillna(75)  # 適当に中間
X_test['cabin_0_50'] = \
    np.where(X_test['cabin_num'] <= 50, 1, 0) # pd.get_dummies( )
X_test['cabin_50_100'] = \
    np.where((50 < X_test['cabin_num']) & (X_test['cabin_num'] <= 100), 1, 0) # pd.get_dummies( ) 
X_test['cabin_100_'] = \
    np.where(100 < X_test['cabin_num'], 1, 0)

# 客室数
X_test['n_cabin'] = X_test['Cabin'].str.count(' ') + 1
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest = RandomForestClassifier(criterion='entropy', n_estimators=4, random_state=1, n_jobs=2)
```
